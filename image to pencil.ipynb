```python
from matplotlib import pyplot as plt
import cv2
from PIL import Image
from IPython.display import display
```

```python
image='flower.jpeg'
x=cv2.imread(image)
display(Image.fromarray(x))
```

```python
image_gray=cv2.cvtColor(x,cv2.COLOR_BGR2GRAY)
display(Image.fromarray(image_gray))
```

```python
img_invert=cv2.bitwise_not(image_gray)
display(Image.fromarray(img_invert))
```

```python
img_smoothing=cv2.GaussianBlur(img_invert,(21,21),sigmaX=0,sigmaY=0)
display(Image.fromarray(img_smoothing))
```

```python
def flw(x,y):
    return cv2.divide(x,255-y,scale=256)
final_img=flw(image_gray,img_smoothing)
display(Image.fromarray(final_img))
```
